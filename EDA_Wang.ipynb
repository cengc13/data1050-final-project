```python
alt.Chart(alt.topo_feature(usa, 'counties')).mark_geoshape(
    stroke='#aaa', strokeWidth=0.25
).transform_lookup(
    lookup='id', from_=alt.LookupData(data = liveCounty, key='fips', fields=['cases'])
).encode(
    alt.Color('cases:Q',
              scale=alt.Scale(domain=[0, 20000], clamp=True), 
              legend=alt.Legend(format='d')),
    alt.Tooltip('cases:Q', format='d')
).project(
    type='albersUsa'
).properties(
    width=900,
    height=500
).configure_view(
    stroke=None
)
```
